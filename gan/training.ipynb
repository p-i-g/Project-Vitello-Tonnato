```python
def apply_augmentations(img):
    # x = KL.RandomFlip('horizontal')(img)
    # x = KL.RandomRotation((-0.15, 0.15))(x)
    # x = KL.RandomZoom(-0.1, 0.1)(x)
    # x = KL.RandomTranslation(0.1, 0.1)(x)
    x = keras.preprocessing.image.random_shear(img[0], 0.15)
    # x = KL.Resizing(286, 286)(x)
    return tf.expand_dims(x, axis=0)

def create_batch(img, batch_size=7):
    imgs = []
    for i in range(batch_size):
        imgs.append(apply_augmentations(img))
    return tf.concat(imgs, axis=0)
```
