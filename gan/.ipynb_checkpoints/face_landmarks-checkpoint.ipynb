```python
@tf.function
def gaussian(shape=(28, 28), peakValue=1.0, center=tf.constant((14.0, 14.0)),
             spread=1.0):  # center[i] between 0-shape[i]
    grid = tf.stack(
        tf.meshgrid(
            tf.range(shape[1], dtype=tf.float32),
            tf.range(shape[1], dtype=tf.float32)
        ),
        axis=-1
    )
    unstacked = tf.reshape(grid, [-1, 2])

    numerator = tf.square(center - unstacked)
    denominator = 2 * tf.square(spread)

    exponent = -tf.reduce_sum(numerator / denominator, axis=-1)

    unstacked_gaussian = peakValue * tf.exp(exponent)
    return tf.reshape(unstacked_gaussian, shape)


plt.imshow(gaussian(shape=[28, 28], peakValue=1.0, center=tf.constant((1, 14.5)), spread=1.0).numpy())
```

```python
@tf.function
def offset(shape=(28, 28), center=tf.constant((0.5, 0.5))):  # center between 0-1
    grid = tf.stack(
        tf.meshgrid(
            tf.range(shape[1], dtype=tf.float32),
            tf.range(shape[1], dtype=tf.float32)
        ),
        axis=-1
    ) / shape[1]
    unstacked = tf.reshape(grid, [-1, 2])
    distances = unstacked - center
    x, y = tf.unstack(distances, axis=-1)
    return tf.reshape(x, shape), tf.reshape(y, shape)
```

```python
plt.imshow(np.abs(offset(center=(0.8, 0.5))[0].numpy()))
```

```python
def gen_data(landmarks, rect, filename):
    x_min, y_min, x_max, y_max = tf.unstack(rect)

    img = tf.io.read_file(WFLW_ROOT + "WFLW_images/" + filename)
    img = tf.image.decode_image(img)
    img = img[y_min: y_max, x_min:x_max, :]
    img = tf.image.resize_with_pad(img, 256, 256)

    landmarks = tf.reshape(landmarks, shape=(-1, 2))
    landmarks = landmarks - tf.cast(tf.stack([x_min, y_min]), tf.dtypes.float32)
    landmarks = landmarks / tf.cast(tf.stack([x_max - x_min, y_max - y_min]), tf.dtypes.float32)

    heatmaps = []
    x_offsets = []
    y_offsets = []

    for coords in tf.unstack(landmarks):
        heatmaps.append(gaussian(shape=(28, 28), peakValue=1.0, center=coords * 28, spread=1.0))
        x_offset, y_offset = offset(center = coords)
        x_offsets.append(x_offset)
        y_offsets.append(y_offset)

    heatmaps = tf.stack(heatmaps, axis=-1)
    x_offsets = tf.stack(x_offsets, axis=-1)
    y_offsets = tf.stack(y_offsets, axis=-1)

    return img, tf.concat([heatmaps, x_offsets, y_offsets], axis=-1)
```

```python
for i in raw_ds.map(gen_data).take(1):
    plt.imshow(i[0].numpy().astype('int'))
```
